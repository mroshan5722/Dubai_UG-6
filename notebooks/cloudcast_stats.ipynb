```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
from tqdm import tqdm
```

```python
# Define the directories containing the .npy files
train_dir = '../data/raw_data/cloudcast/CloudCastSmall/TrainCloud'
test_dir = '../data/raw_data/cloudcast/CloudCastSmall/TestCloud'

print("train data:", len([f for f in os.listdir(train_dir) if f.endswith('.npy')]))
print("test data:", len([f for f in os.listdir(test_dir) if f.endswith('.npy')]))
# List all .npy files from both directories
npy_files = [f for f in os.listdir(train_dir) if f.endswith('.npy')] + \
            [f for f in os.listdir(test_dir) if f.endswith('.npy')]

print("total data:", len(npy_files))
# Print the combined list of .npy files
print("Combined list of .npy files:")
print(npy_files)
```

```python
# Load the .npy files
data_list = []  # List to store loaded data
for file in npy_files:
    # Determine the full path based on the file's origin
    if file in os.listdir(train_dir):
        data = np.load(os.path.join(train_dir, file))
    else:
        data = np.load(os.path.join(test_dir, file))
    
    data_list.append(data)  # Append loaded data to the list
```

```python
data_stats = []
for file in npy_files:
    # data = np.load(os.path.join('your_directory', file))
    
    stats = {
        'filename': file,
        'shape': data.shape,
        'dtype': data.dtype,
        'min': np.min(data),
        'max': np.max(data),
        'mean': np.mean(data),
        'std': np.std(data),
        'has_nan': np.isnan(data).any(),
        'has_inf': np.isinf(data).any()
    }
    
    data_stats.append(stats)


# print the loaded data shapes
# for i, data in enumerate(data_list):
#     print(f"Loaded data shape for file {npy_files[i]}: {data.shape}")

# Print the statistics for each file
for stat in data_stats:
    print(stat)
```
